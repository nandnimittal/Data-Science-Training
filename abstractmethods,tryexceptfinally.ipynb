```python
from abc import ABC, abstractmethod
```

```python
class employee(ABC):
    @abstractmethod
    def calculate_salary(self):
        pass
    @abstractmethod
    def details(self):
        pass
```

```python
class employees(employee):
    def salary(self):
        print('this is salary')#as here we didn't implement calculate_salary and details function given in the employee class
```

```python
user1=employees()
```

```python
class employees1(employee):
    def __init__(self):
        self.name = input('enter  name')
        self.salary = int(input('enter salary'))
        self.pf = self.salary*0.20
        self.calculate_salary()# we call this function in constructor as if we make the object then this init will be called 
        #then if we call the details function then it gives error as in details function we need net_salary 
        #so for that we need to call the calculate_salary function always before details function
        #so to avoid repeatedly call of calculate_salary function so we call in the constructor so that we call always when the new object is formed
    def calculate_salary(self):
        self.net_salary = self.salary + self.pf
        print(self.net_salary)
    def details(self):
        print(f"name of the employye is {self.name} and its salary is {self.net_salary}")
```

```python
user2=employees1()
```

```python
user2.calculate_salary()
```

```python
user2.details()
```

```python
#exception handling - try and except, finally
```

```python
a=int(input('num1-'))
b=int(input('num2-'))
print(a/b)#gives the error
```

```python
try:
    a=int(input('num1-'))
    b=int(input('num2-'))
    print(a/b)
except:
    print('number cant be divide by 0')
```

```python
try:
    a=int(input('num1-'))
    b=int(input('num2-'))
    print(a/b)
except (ZeroDivisionError):
    print('number cant be divide by 0')
except (ValueError):#if we entered the value of 'a' or 'b' as string
    print('enter number only')
```

```python
try:
    a=int(input('num1-'))
    b=int(input('num2-'))
    print(a/b)
except (ZeroDivisionError):
    print('number cant be divide by 0')
except (ValueError):
    print('enter number only')
finally:
    print('program end successfully')
```

```python
try:
    a=int(input('num1-'))
except (ValueError):
    print('taking default number')
    a=0
try:
    b=int(input('num2-'))
except (ValueError):
    print('taking default number')
    b=0
try:
    print(a/b)
except (ZeroDivisionError):
    print('number cant be divide by 0')
    print('mul-', a*b)
    print('sum', a+b)
```

```python
#so last two print statements are not executed as if the input doesn't enters except block so we use finally for that
try:
    a=int(input('num1-'))
except (ValueError):
    print('taking default number')
    a=0
try:
    b=int(input('num2-'))
except (ValueError):
    print('taking default number')
    b=0
try:
    print(a/b)
except (ZeroDivisionError):
    print('number cant be divide by 0')
finally:
    print('mul-', a*b)
    print('sum', a+b)
```
