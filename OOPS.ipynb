```python
# bankaccount code
class bankaccount:
    def __init__(self, name, balance = 0):
        self.name = name 
        self.balance = balance
        self.transactions = []
        self.transactions.append(f"Name:- {name}")
        self.transactions.append(f"opening balance = {balance}")
        print('welcome to our bank')
    def deposit(self,amount):     
         if amount > 0:
             self.balance += amount
         else:
             print('amount must be in positive')
             self.transactions.append(f" deposit amount is {amount} and balance is {self.balance}")
    def withdrawl(self,amount):
          if amount<self.balance:
              self.balance -= amount
              self.transactions.append(f"withdraw amount:- {amount} balance = {self.balance}")
          else:
              print('insufficient balance')
    def viewbalance(self):
          print(f" {self.name},{self.balance}")    
    def transaction(self):          
          for i in self.transactions:
              print(i)
                 
    
```

```python
class BankAccount:
    users = {}

    def __init__(self, username, password, balance=0):
        self.username = username
        self.password = password
        self.balance = balance
        self.transactions = []

    @classmethod
    def register(cls):
        u = input("Create username: ")
        p = input("Create password: ")

        # simple strong check
        if len(p) < 6 or p.islower() or p.isalpha():
            print("Weak password! Use letters + numbers.")
            return

        cls.users[u] = BankAccount(u, p)
        print("Account created!\n")

    @classmethod
    def login(cls):
        u = input("Username: ")
        p = input("Password: ")

        if u in cls.users and cls.users[u].password == p:
            print("Login successful!\n")
            return cls.users[u]
        else:
            print("Wrong details!\n")
            return None

    def deposit(self, amt):
        self.balance += amt
        self.transactions.append(f"Deposited {amt}")

    def withdraw(self, amt):
        if amt <= self.balance:
            self.balance -= amt
            self.transactions.append(f"Withdrew {amt}")
        else:
            print("Insufficient balance")

    def show(self):
        print("Balance:", self.balance)
        print("Transactions:", self.transactions)


# -------- Main --------
while True:
    print("\n1.Register  2.Login  3.Exit")
    ch = input("Choice: ")

    if ch == '1':
        BankAccount.register()

    elif ch == '2':
        user = BankAccount.login()
        if user:
            while True:
                print("\n1.Deposit 2.Withdraw 3.Show 4.Logout")
                op = input("Option: ")

                if op == '1':
                    user.deposit(int(input("Amount: ")))
                elif op == '2':
                    user.withdraw(int(input("Amount: ")))
                elif op == '3':
                    user.show()
                elif op == '4':
                    break

    elif ch == '3':
        break
```
