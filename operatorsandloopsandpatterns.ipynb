```python
# inverse right angle triangle : phele 4 times space print krega and 1 time 
n=5
for i in range(1,n+1):
    for space in range(n-i):
        print(' ',end=' ')
    for star in range(i):
        print('*',end=' ')
    print()    
```

```python
reverse inverse right angle triangle.
n=5
for i in range(n,0,-1):
    for space in range(n-i):
        print(' ',end=' ')
    for star in range(i):
        print('*',end=' ')
    print()  
```

```python
# reverse pyrimad
n = 5 
for i in range(n,0,-1):
    print(" "*(n-i), end= ' ')
    print("* "*i)
```

```python
# reverse a diamond
n = 5 
m=4
for i in range(1 , n + 1):
    print(" "*(n-i), end= ' ')
    print("* "*i)
for i in range(m,0,-1):
    print(" "*(n-i), end= ' ')
    print("* "*i)    
```
