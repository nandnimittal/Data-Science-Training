```python
#challenge in method2 - it gives 1ef, NaN, inf as output because in python these all are numbers here e means 1000
list4 = ['1e3', '3','4','NaN','inf']
nums2=[]
for i in list4:
    try:
        nums2.append(float(i))
    except:
        pass
print(nums2)
```

```python
# to solve this
list5 = ['1e3', '3','4','NaN','inf']
nums3=[]
for i in list5:
    try:
        val = float(i)
        if not (val == float('1e3') or val == float('inf') or val == float('-inf') or val != val):
                nums3.append(val)
    except:
        pass
print(nums3)
```

```python
#example 1
expenses = np.array([
    [200, 340, 230, 150, 450], #food
    [100, 350, 120, 600, 244], #travel
    [300, 450, 256, 800, 679]  #shopping
])
```

```python
print('total expense: ',np.sum(expenses))# these operations are very fast
print('category wise sum: ',np.sum(expenses, axis =1))
print('day wise total: ',np.sum(expenses, axis =0))
print('highest expense day: ', np.max(np.sum(expenses, axis=0)))
```

```python
#example 2
marks = np.array([34,56,65,78,87,65,66])
```

```python
#where if like if else statement - takes less time- fast
grades =np.where(marks >= 75, "A", 
        np.where(marks >= 55, "B",
        np.where(marks >= 40, "C",  "Fail")))
```

```python
print(grades)
```
